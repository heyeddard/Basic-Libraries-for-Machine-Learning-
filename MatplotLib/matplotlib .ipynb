```python
x = np.linspace(0,10,20)
y=x*2
plt.plot(x,y,'b+')
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("OHM'sLAW")
plt.show()
```

```python
x = np.linspace(0,5,10)
y = np.linspace(3,6,10)
plt.plot(x,y,'r-',x,y**2,'b+',x,y**3,'g^')
plt.show()
```

```python
x = np.linspace(1,10,100)
y = np.log(x)
plt.figure(1)
plt.subplot(121)# 1 row 2 column start with 1 graph
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(122)# 1 row 2 column start with 1 graph
plt.title("y=log(x)**2")
plt.plot(x,y**2)
plt.show()


```

```python
x = np.linspace(1,10,100)
y = x
plt.figure(1)
plt.subplot(221)# 1 row 2 column start with 1 graph
plt.title("linear")
plt.plot(x,y)

plt.subplot(222)# 1 row 2 column start with 1 graph
plt.title("cubic")
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(221)# 1 row 2 column start with 1 graph
plt.title("log")
plt.plot(x,np.log(x))


plt.subplot(222)# 1 row 2 column start with 1 graph
plt.title("exponential")
plt.plot(x,x**2)
plt.show()


```
